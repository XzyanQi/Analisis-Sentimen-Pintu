# B. Scrapping Data

```python
# Mengimpor pustaka google_play_scraper untuk mengakses ulasan dan informasi aplikasi dari Google Play Store.
from google_play_scraper import app, reviews, Sort

# Mengambil 30000 ulasan dari aplikasi dengan ID 'com.valar.pintu' di Google Play Store.
scrapreview, continuation_token = reviews(
    'com.valar.pintu',            # ID aplikasi Pintu
    lang='id',                    # Bahasa ulasan: Indonesia
    country='id',                 # Negara: Indonesia
    sort=Sort.MOST_RELEVANT,      # Urutan ulasan: Rekevant
    count=30000                   # Jumlah maksimum ulasan yang ingin diambil
)
```

```python
# Menyimpan ulasan dalam file CSV
import csv

with open('ulasan_pintu.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Review'])  # Menulis header kolom
    for review in scrapreview:
        writer.writerow([review['content']])  # Menulis konten ulasan ke dalam file CSV
```

```python
import pandas as pd  # Pandas untuk manipulasi dan analisis data
pd.options.mode.chained_assignment = None  # Menonaktifkan peringatan chaining

app_review_df = pd.DataFrame(scrapreview)
app_review_df.shape
app_review_df.head()
app_review_df.to_csv('ulasan_pintu.csv', index=False)


# Membuat DataFrame dari hasil scrapreview
app_review_df = pd.DataFrame(scrapreview)

# Menghitung jumlah baris dan kolom dalam DataFrame
jumlah_ulasan, jumlah_kolom = app_review_df.shape
```

```python
# Hapus baris dengan nilai NaN
clean_df = app_review_df.dropna()

# Hapus kolom yang tidak dibutuhkan
clean_df = clean_df.drop(['reviewId', 'userName', 'userImage','thumbsUpCount',
                          'reviewCreatedVersion', 'replyContent', 'repliedAt',
                          'appVersion','at'], axis=1)
# Munculin hasil
clean_df
```

```python
# Menghapus indeks
df_pemodelan = clean_df.reset_index(drop=True)
```
